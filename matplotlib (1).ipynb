```python
commoners=['Mr.','Miss.','Mrs.','Master.','Ms.','Mlle.','Mme.','Dona.','Don.']
nobility=['Jonkheer.','Countess.','Lady.']
army_navy=['Major.','Capt.','Sir.','Col.']
religion_medicine=['Rev.','Dr.']
```

```python
len(commoners)+len(nobility)+len(army_navy)+len(religion_medicine)
```

```python
encoder=[]
for titles in titles_df.title:
    if titles in commoners:
        encoder.append("commoners")
    elif titles in nobility:
        encoder.append("royalty")
    elif titles in army_navy:
        encoder.append("military")
    else:
        encoder.append("religion_medicine")
```

```python
titles_df["categories"]=encoder
```

```python
titles_df
```

```python
final_df=titles_df.groupby(['categories']).sum()
```

```python
final_df
```

```python
labels=list(final_df.index)
```

```python
kids = t2[t2['age']<=18]['fare']
youth = t2[(t2['age']>=18)&(t2['age']<=35)]['fare']
mid_age = t2[(t2['age']>35)&(t2['age']<=60)]['fare']
old = t2[t2['age']>60]['fare']
```

```python
mid_age
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10,7))
plt.boxplot([kids,youth,mid_age,old],
            labels=['kids','youth','mid_age','old'],
            medianprops={'linewidth':2})
plt.title("FARE DISTRIBUTION BASED ON AGE")
plt.xlabel("Age Groups")
plt.ylabel("Fare Distribution")
plt.show()
```
